```python
#匯入在sklearn.linear_model套件裡面的LinearRegression模型
#匯入在sklearn.model_selection套件裡面的train_test_split模組
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
#切割數據集(訓練資料80%、測試資料20%，設定random_state=20)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 20)
#畫出訓練資料集的matplotlib圖形m
plt.scatter(x_train, y_train)
print(x_train.shape)
```

## 用訓練資料來 fit 函數
1. 只用訓練資料集的資料進行linear regression演算法<br>
2. 並計算出訓練階段的MSE
3. 畫出目標點(藍色)與預測點(紅色)的對應圖
![](images/LR_train.PNG)

```python
myLnrg02 = LinearRegression()
x_train_rs = x_train.reshape(-1,1)
myLnrg02.fit(x_train_rs, y_train)
y_train_pred = myLnrg02.predict(x_train_rs)
MseTrain = round(np.sum((y_train_pred - y_train)**2)/len(y_train), 6)
MseTrain
```

```python
plt.scatter(x_train, y_train)
plt.plot(x_train, y_train_pred, c='r')
```

## 將訓練出來的函數預測測試集的X值
1. 使用剛剛訓練出來的模型進行測試資料集的資料預測*注意reshape<br>
使用X = 2.44897959，預測出來數值應該為 4.3025375<br>
所有測試集資料
$$\widehat{y}=xw+b=w_{1}x_{1}+b$$

```python
x_test_rs = x_test.reshape(-1,1)
y_test_pred = myLnrg02.predict(x_test_rs)
print(y_test_pred)
```

```python
w = myLnrg02.coef_
b = myLnrg02.intercept_
w * 2.44897959 + b
```

2. 並計算出測試階段的MSE

```python
MseTest = round(np.sum((y_test_pred - y_test)**2)/len(y_test), 6)
MseTest
```

3. 畫出目標點(藍色)與預測點(紅色)的對應圖
![](images/LR_test.PNG)

```python
myTitle = 'f(x) = 1.6x + 0.2, mse:' + str(round(MseTest, 2))
plt.title(myTitle)
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x_test, y_test, alpha=0.8, label='Ground Truth')
plt.scatter(x_test, y_test_pred, c='r', alpha=0.8,label='Predict')
plt.legend()
```
