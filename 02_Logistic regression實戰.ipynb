```python
#載入套件
import pandas as pds
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#使用資料夾內鐵達尼號的資料集，練習邏輯迴歸的處理邏輯
train = pds.read_csv('data/titanic_data.csv')
train.head(10)
```

```python
#探索資料
# 用heatmap(.isnull()來找出缺失的資料在哪些欄位。Age似乎約有五分之一的資料缺失，可用合理資料補充。)
sns.heatmap(train.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
# Embarked 也有 Missing，用眾數補值
print("Missing Embarked Count:", len(train) - len(train["Embarked"].dropna()))
print(train.groupby(["Embarked"]).size())
```

```python
train["Embarked"].fillna(train["Embarked"].mode()[0], inplace = True)
print(train.groupby(["Embarked"]).size())
print("Missing Embarked Count:", len(train) - len(train["Embarked"].dropna()))
```

```python
#用countplot來看性別是否影響最後存活。女性存活比例比男性高。
sns.countplot(x='Survived',hue='Sex',data=train,palette='RdBu_r')
```

```python
#艙等是否也影響存活?
sns.countplot(x='Survived',hue='Pclass',data=train)
```

```python
#用直方圖看年齡分佈。缺失資料在此不計。
sns.distplot(train['Age'].dropna(),kde=False,bins = 30)
# sns.distplot(train['Age'],kde=False,bins=30)
```

```python
#SibSp代表兄弟姐妹或眷屬人數
sns.countplot(x='SibSp', data=train)
```

```python
#用直方圖看票價分佈
train['Fare'].hist(bins = 50, figsize=(10,4))
```

```python
#資料清理
#剛有看到年齡資料約有五分之一缺失，一種做法是棄之不用，一種可以考慮把年齡平均放進去。
#使用盒鬚圖來看年齡和艙等的分佈。看起來頭等艙年齡較高，隨之降低。
plt.figure(figsize=(12,6))
my_pal = {"1": "y", "2": "b", "3":"m"}
sns.boxplot(x='Pclass',y='Age', data=train, palette=my_pal)
```

```python
train.describe()
```

```python
# 把艙等平均年齡填入缺失的資料中
# 艙等1平均年齡是37，艙等2平均年齡是29，艙等3平均年齡是24 (用程式計算較正確的數字，再用 Dict 帶入)
# 用 dict 的方式取代

AvgAgebyPclass = pds.DataFrame(train.groupby(["Pclass"])["Age"].mean().round(0))
AvgAgebyPclassDict = pds.Series(AvgAgebyPclass.Age, index = AvgAgebyPclass.index).to_dict()

def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    if pds.isnull(Age):
        return AvgAgebyPclassDict.get(Pclass)       
        # if Pclass == 1:
        #     return 37
        # elif Pclass == 2:
        #     return 29
        # else:
        #     return 24
    else:
        return Age

train['Age'] = train[['Age','Pclass']].apply(impute_age, axis=1)
```

```python
train.describe()
```

```python
#用heatmap查看年齡是否還有缺失資料
sns.heatmap(train.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
train.dropna(inplace=True) #inplace： False: 返回新的數據集（默認） True: 在原數據集上操作
```

```python
#虛擬變數範例
#使用pandas(d.get_dummies)將性別從類別資料轉為數值變數。
sex = pd.get_dummies(train['Sex'], drop_first=True, dtype = int)
#請將drop_first=True，如果為True會刪除第一個類別值。
#本來若有k個level，drop_first=True會只剩下k-1個level，降低共線性的可能
sex.head()
```

```python
#使用pandas將上船的港口從類別資料改換為數值變數(drop_first=True)
embark = pd.get_dummies(train['Embarked'], drop_first=True, dtype = int)
embark.head()
```

```python
#將性別及上船港口的數值變數和原始的data結合
train = pd.concat([train,sex,embark],axis=1)
train.head()
```

```python
#將原始的性別、登船港口、票號、姓名欄位棄用
train.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
train.head()
```

```python
#因為PassengerId在此也只是排序，因此也棄用
train.drop('PassengerId',axis=1,inplace=True)
train.head()
```

```python
#建立feature資料集(選取survived以外所有欄位)與target資料集(survived)
X = train.drop('Survived',axis=1)
y = train['Survived']
```

```python
#載入訓練組及測試組分組套件
from sklearn.model_selection import train_test_split
```

```python
#設定隨機種子(random_state) 為101，測試組比例佔30%
X_train, X_test, y_train, y_test = train_test_split(X, y ,test_size=0.3, random_state=101)
```

```python
#載入邏輯回歸
from sklearn.linear_model import LogisticRegression
```

```python
#建立訓練組邏輯迴歸模型
LogisticModel = LogisticRegression()
LogisticModel.fit(X_train,y_train)
```

```python
#使用模型進行測試集資料的預測
predictions = LogisticModel.predict(X_test)
predictions2 = LogisticModel.predict(X_train)
```

```python
#ground truth
print(y_test.values)
print(y_train.values)
```

#### 計算accuracy及結果顯示
accuracy<br>
error_index #錯誤答案是哪幾筆資料<br>
miss classification #錯誤答案<br>
錯誤欄位的正確之原始資料表格

```python
error_index2=np.where(predictions2!=y_train.values)[0]
print('train accuracy:',1-len(error_index2)/len(X_train))
error_index=np.where(predictions!=y_test.values)[0]
print('test accuracy:',1-len(error_index)/len(X_test))
```

```python
print('error_index:',error_index)
print('miss classification :',predictions[error_index]) #錯誤答案
data_index=train.iloc[error_index].index                #這裡變數 train 是所有資料
train.iloc[data_index]                                  #錯誤欄位的正確之原始資料
```
